```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Q1: Create two charts using list and dataframe.

```python
names = ['Hajar', 'Sara', 'Ghadi', 'Joree']
ages = [22, 18, 19, 21]

# Create bar chart
plt.bar(names, ages,color='violet')
plt.xlabel('Name')
plt.ylabel('Age')
plt.title('Age Distribution')
plt.show()
```

```python
data = {'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May'],
        'Revenue': [1000, 1200, 900, 1500, 1800],
        'Profit': [200, 300, 100, 400, 500]}

df = pd.DataFrame(data)

# Create line chart
plt.plot(df['Month'], df['Revenue'], label='Revenue')
plt.plot(df['Month'], df['Profit'], label='Profit')

plt.xlabel('Month')
plt.ylabel('Amount ')
plt.title('Revenue and Profit Trend')

plt.legend()
plt.show()
```

## Q2: Create pie chart with equal explode values.

```python
pieLabels = 'Asia', 'Africa', 'Europe', 'North America', 'South America'
# Population data
populationShare= [59.69, 16, 9.94, 7.79, 5.68]
colors = ['#ff9999','#66b3ff','#99ff99', '#ffcc39']
plt.title('Population Distribution',fontsize = 15)
plt.pie(populationShare, labels=pieLabels, autopct='%1.1f%%',  startangle=180, radius=1.5,explode = [0,0,0,0,0])

plt.show()
```

## Q3: Create a scatter chart with circle marker, alpha = 0.4, and size = 50.

```python
plt.figure()
plt.scatter(names, ages,alpha = 0.4, s=50)
plt.xlabel('Name')
plt.ylabel('Age')
plt.title('Age Distribution')
plt.show()
```

## Q4: Add the following to your chart:
- title placed in the left and has a red color
- x and y label with font size = 16

```python
plt.figure()
plt.scatter(names, ages,alpha = 0.4, s=50)
plt.title('Age vs. Name',fontsize = 20, loc='left', color='red')
plt.xlabel('Name',fontsize = 16)
plt.ylabel('Age',fontsize = 16)
plt.show()
```

## Q5: Create three line charts with Legends at left and colors.

```python
x = [1, 2, 3, 4, 5]
y1 = [10, 15, 12, 8, 11]
y2 = [5, 9, 6, 4, 7]
y3 = [18, 13, 16, 9, 14]
plt.plot(x, y1, label='curve  1', color='purple')
plt.plot(x, y2, label='curve  2', color='orange')
plt.plot(x, y3, label='curve  3', color='green')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Line Charts')
# Customize legend
plt.legend(loc='upper left')

plt.show()
```

## Q6: Create subplots with three charts and figsize=20,10.

```python
fig = plt.figure(figsize=(20,10))
fig.add_subplot(1,3,1)
plt.pie(populationShare, labels=pieLabels, radius=1.5)
fig.add_subplot(2,3,2)
plt.scatter(names, ages,alpha = 0.4, s=50)
plt.title('Age vs. Name',fontsize = 20, loc='left', color='red')
plt.xlabel('Names',fontsize = 16)
plt.ylabel('Ages',fontsize = 16)
fig.add_subplot(2,3,3)
plt.plot(x, y1, label='curve  1', color='purple')
plt.plot(x, y2, label='curve  2', color='orange')
plt.plot(x, y3, label='curve  3', color='green')
plt.legend(loc='upper left')
plt.show()
```
